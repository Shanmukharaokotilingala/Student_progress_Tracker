**STUDENT PROGRESS REPORT**

```python
print("-----------------------------------")
print("  STUDENT PROGRESS TRACKER SYSTEM  ")
print("-----------------------------------")

name = input("Enter Student Name: ")
cls = input("Enter Class: ")
roll = input("Enter Roll Number: ")
count = int(input("How many subjects? "))
total = 0
marks = []
for i in range(1, count + 1):
    mark = float(input(f"Enter marks of subject {i}: "))
    marks.append(mark)
    total = total + mark
avg = total / count
if avg >= 90:
    grade = "A"
elif avg >= 75:
    grade = "B"
elif avg >= 50:
    grade = "C"
elif avg >= 35:
    grade = "D"
else:
    grade = "Fail"
result = "PASS"
for m in marks:
    if m < 35:
        result = "FAIL"
        break
print("-----------------------------------")
print("        STUDENT PROGRESS REPORT     ")
print("-----------------------------------")
print("Name        :", name)
print("Class       :", cls)
print("Roll No     :", roll)
print("Total Marks :", total)
print("Average     :", avg)
print("Grade       :", grade)
print("Result      :", result)
print("-----------------------------------")
last_name = name
last_class = cls
last_roll = roll
last_total = total
last_avg = avg
last_grade = grade
last_result = result
while True:
    print("\n1. Enter marks again")
    print("2. View last report")
    print("3. Save report and exit")
    print("4. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        name = input("Enter Student Name: ")
        cls = input("Enter Class: ")
        roll = input("Enter Roll Number: ")
        count = int(input("How many subjects? "))
        total = 0
        marks = []
        for i in range(1, count + 1):
            mark = float(input(f"Enter marks of subject {i}: "))
            marks.append(mark)
            total = total + mark
        avg = total / count
        if avg >= 90:
            grade = "A"
        elif avg >= 75:
            grade = "B"
        elif avg >= 50:
            grade = "C"
        elif avg >= 35:
            grade = "D"
        else:
            grade = "Fail"
        result = "PASS"
        for m in marks:
            if m < 35:
                result = "FAIL"
                break
        print("-----------------------------------")
        print("        STUDENT PROGRESS REPORT     ")
        print("-----------------------------------")
        print("Name        :", name)
        print("Class       :", cls)
        print("Roll No     :", roll)
        print("Total Marks :", total)
        print("Average     :", avg)
        print("Grade       :", grade)
        print("Result      :", result)
        print("-----------------------------------")
        last_name = name
        last_class = cls
        last_roll = roll
        last_total = total
        last_avg = avg
        last_grade = grade
        last_result = result
    elif choice == "2":
        print("-----------------------------------")
        print("       LAST SAVED STUDENT REPORT    ")
        print("-----------------------------------")
        print("Name        :", last_name)
        print("Class       :", last_class)
        print("Roll No     :", last_roll)
        print("Total Marks :", last_total)
        print("Average     :", last_avg)
        print("Grade       :", last_grade)
        print("Result      :", last_result)
        print("-----------------------------------")
    elif choice == "3":
        filename = f"{last_name}.txt"
        f = open(filename, "w")
        f.write("STUDENT PROGRESS REPORT\n")
        f.write("---------------------------\n")
        f.write(f"Name        : {last_name}\n")
        f.write(f"Class       : {last_class}\n")
        f.write(f"Roll No     : {last_roll}\n")
        f.write(f"Total Marks : {last_total}\n")
        f.write(f"Average     : {last_avg}\n")
        f.write(f"Grade       : {last_grade}\n")
        f.write(f"Result      : {last_result}\n")
        f.close()
        print("Report saved. Exiting...")
        break
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice!")
```
